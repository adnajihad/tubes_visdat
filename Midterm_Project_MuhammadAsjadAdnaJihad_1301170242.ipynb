Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
df1 = pd.read_csv('/content/tubes_visdat/Nasdaq.csv', parse_dates=True)
df1['Date'] = pd.to_datetime(df1['Date'])
df1['year']= df1['Date'].dt.year
df1_target = df1[df1['year']== 2019]

df1_target['Change'] = df1_target.Close.pct_change() * 100

df1_target
```

```python
import numpy as np
df1_target['Trend'] = np.nan


df1_target['Trend'] = np.where(df1_target['Change'] <= -7, 'Bear Drop', df1_target['Trend'])
df1_target['Trend'] = np.where((df1_target['Change'] > -7) & (df1_target['Change'] <= -3), 'Among Top Losers', df1_target['Trend'])
df1_target['Trend'] = np.where((df1_target['Change'] > -3) & (df1_target['Change'] <= -1), 'Negative', df1_target['Trend'])
df1_target['Trend'] = np.where((df1_target['Change'] > -1) & (df1_target['Change'] <= -0.5), 'Negative', df1_target['Trend'])
df1_target['Trend'] = np.where((df1_target['Change'] > -0.5) & (df1_target['Change'] <= 0.5), 'Slight or No Change', df1_target['Trend'])
df1_target['Trend'] = np.where((df1_target['Change'] > 0.5) & (df1_target['Change'] <= 1), 'Slight Positive', df1_target['Trend'])
df1_target['Trend'] = np.where((df1_target['Change'] > 1) & (df1_target['Change'] <= 3), 'Positive', df1_target['Trend'])
df1_target['Trend'] = np.where((df1_target['Change'] > 3) & (df1_target['Change'] <= 7), 'Among Top Gathers', df1_target['Trend'])
df1_target['Trend'] = np.where(df1_target['Change'] > 7, 'Bull Run', df1_target['Trend'])

df1_target
```

```python
plt_df1 = sns.countplot(y="Trend", data=df1_target)

plt_df1.set_xticklabels(plt_df1.get_xticklabels(), rotation=40, ha="right")
total = len(df1_target['Trend'])
for p in plt_df1.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        plt_df1.annotate(percentage, (x, y))
plt.title('NIKKEI 2019 TREND ANALYSIS')
plt.tight_layout()
plt.show()
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
clt_df1 = df1_target['Change']
clt_df2 = df2_target['Change']
clt_df3 = df3_target['Change']

dftemp = df3_target

dftemp['NIKKEI'] = clt_df1
dftemp['NASDAQ'] = clt_df2
dftemp['HANGSENG'] = clt_df3

df_change = dftemp[['Date','NIKKEI','NASDAQ','HANGSENG']]

df_change.dropna(axis = 0, inplace = True)

df_change
```

```python
#Memvisualiasikan volatility
vol = df_change[['NIKKEI', 'NASDAQ', 'HANGSENG']].rolling(7).std()*np.sqrt(7)
vol.plot(figsize = (15, 7))
plt.xlabel('NIKKEI, NASDAQ, & HANGSENG STOCK VOLATILITY IN 2019')
```

#### Penjelasan 
berdasarkan visualisasi data diatas, dapat dilihat grafik volatilitas saham ketiga dataset dalam tahun 2019 dan dapat ditarik kesimpulan dimana kenaikan tertinggi terjadi pada perusahaan NIKKEI dan penurunan terbesar terjadi pada perusahaan NASDAQ.
